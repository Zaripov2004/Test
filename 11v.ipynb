1️⃣ Bosqich: Loyiha nomi va maqsadi

Nom: Dunyo transport statistikasi (2010–2024)
Maqsad: Havo, avtomobil va temir yo‘l transportining global hajmlarini o‘rganish, trendlarni tahlil qilish va kelgusiga prognoz qilish.

2️⃣ Bosqich: Ma’lumotlarni olish

World Bank indikatorlari:

IS.AIR.PSGR – Havo yo‘lovchilari soni (million)

IS.ROD.PAVE.KM – Avtomobil yo‘llari uzunligi (km)

IS.RRS.PASG.KM – Temir yo‘l yo‘lovchilari (yo‘lovchi-km)

IATA: Yo‘lovchi tashish statistikasi

ITF: Transport oqimlari

```python
import pandas as pd

data = {
    "Year": [2018, 2019, 2020, 2021, 2022, 2023],
    "Air_Passengers": [4200, 4400, 1900, 2500, 3700, 4200],  # mln yo‘lovchi
    "Road_Length": [33000000, 33200000, 33400000, 33600000, 33800000, 34000000],  # km
    "Rail_Passenger_KM": [2800, 2850, 2300, 2400, 2600, 2700]  # mlrd yo‘lovchi-km
}

df = pd.DataFrame(data)
df
```

3️⃣ Bosqich: Ma’lumotlarni tayyorlash

```python
df["Year"] = df["Year"].astype(int)
df.describe()
```

4️⃣ Bosqich: Havo, avtomobil va temir yo‘l transportini chizma orqali taqqoslash

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(df["Year"], df["Air_Passengers"], marker="o", label="Havo transporti (mln yo‘lovchi)")
plt.plot(df["Year"], df["Road_Length"]/1000, marker="o", label="Avtomobil yo‘llari (ming km)")
plt.plot(df["Year"], df["Rail_Passenger_KM"], marker="o", label="Temir yo‘l (mlrd yo‘lovchi-km)")
plt.title("Dunyo transport statistikasi (2018–2023)")
plt.xlabel("Yil")
plt.ylabel("Transport ko‘rsatkichlari")
plt.legend()
plt.grid(True)
plt.show()
```

5️⃣ Bosqich: Pandemiya ta’siri (2020-yil)

2020-yilda havo transportida keskin pasayish kuzatiladi.

Temir yo‘l ham kamaygan, ammo avtomobil yo‘llari barqaror o‘sishda qolgan.

6️⃣ Bosqich: Transport turlarini ulushga ajratish (Pie Char)

```python
latest = {
    "Havo": df["Air_Passengers"].iloc[-1],
    "Yo‘l uzunligi (km)": df["Road_Length"].iloc[-1],
    "Temir yo‘l": df["Rail_Passenger_KM"].iloc[-1]
}

plt.pie(latest.values(), labels=latest.keys(), autopct="%1.1f%%")
plt.title("2023-yilda transport turlari taqsimoti")
plt.show()
```

7️⃣ Bosqich: Davlatlar bo‘yicha taqqoslash (demo)

```python
countries = {
    "AQSh": [900, 6800],
    "Xitoy": [1100, 5200],
    "Rossiya": [120, 2100],
    "Yaponiya": [450, 2900],
    "Germaniya": [380, 2500]
}

df_countries = pd.DataFrame(countries, index=["Havo yo‘lovchilari (mln)","Temir yo‘l (mlrd yo‘lovchi-km)"]).T
df_countries.plot(kind="bar", figsize=(9,6))
plt.title("Davlatlar bo‘yicha transport (2023)")
plt.ylabel("Ko‘rsatkichlar")
plt.show()
```

8️⃣ Bosqich: O‘sish trendlari (line fit)

```python
plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Air_Passengers"], label="Havo")
plt.plot(df["Year"], df["Rail_Passenger_KM"], label="Temir yo‘l")
plt.plot(df["Year"], df["Road_Length"]/1000, label="Yo‘l uzunligi (ming km)")
plt.legend()
plt.title("Transport o‘sish trendi (2018–2023)")
plt.show()
```

9️⃣ Bosqich: Korrelatsiya tahlili

1️⃣1️⃣ Bosqich: Geografik vizualizatsiya (Choropleth)

```python
import plotly.express as px

geo_df = pd.DataFrame({
    "Country":["China","United States","Russia","Japan","Germany"],
    "Air_Passengers":[1100,900,120,450,380]
})

fig = px.choropleth(
    geo_df,
    locations="Country",
    locationmode="country names",
    color="Air_Passengers",
    title="Havo yo‘lovchilari soni (2023, mln)"
)
fig.show()
```

1️⃣2️⃣ Bosqich: Prognoz qilish (Linear Regression – havo transporti)

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = df[["Year"]]
y = df["Air_Passengers"]

model = LinearRegression()
model.fit(X, y)

future_years = np.array(range(2024, 2030)).reshape(-1,1)
forecast = model.predict(future_years)

plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Air_Passengers"], marker='o', label="Haqiqiy")
plt.plot(future_years, forecast, marker='x', linestyle="--", label="Prognoz")
plt.title("Dunyo havo yo‘lovchilari prognozi (mln)")
plt.xlabel("Yil")
plt.ylabel("Yo‘lovchilar (mln)")
plt.legend()
plt.show()
```
