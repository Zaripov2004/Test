## 4


```python
# отберем признаки с наиболее высокой корреляцией с целевой переменной
# и поместим их в переменную X
X = aholi_df[['2010', "2012"]]
#X = boston_df[['RM', 'LSTAT']]
```

```python
y = aholi_df['2024']
```

```python
print(type(X), type(y))
```

```python
from sklearn.model_selection import train_test_split

# разобьем данные на обучающую и тестовую выборку
# размер тестовой выборки составит 30%
# также зададим точку отсчета для воспроизводимости
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.3,
                                                    random_state = 42)
```

```python
print(X_train.shape, y_train.shape)

# и тестовой выборки
print(X_test.shape, y_test.shape)
```

## 5


```python
from sklearn.linear_model import LinearRegression

# создадим объект этого класса и запишем в переменную model
model = LinearRegression()

# обучим модель
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

# выведем первые пять значений с помощью диапазона индексов
print(y_pred[:])
```

```python
# импортируем функцию корня среднеквадратической ошибки
from sklearn.metrics import root_mean_squared_error

# сравним тестовые и прогнозные значения цен на жилье
print('Root Mean Squared Error (RMSE):', root_mean_squared_error(y_test, y_pred))
```

```python
from sklearn.metrics import r2_score
print('R2:', np.round(r2_score(y_test, y_pred), 2))
```

## 6

```python
r2_score(y_test, y_pred), model.score(X_test, y_test)
```
